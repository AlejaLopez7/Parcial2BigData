```python
# Importing required libraries
import time
from bs4 import BeautifulSoup
from selenium import webdriver
from pymongo import MongoClient
```

```python
# Base URL of the site to be analyzed
SITE_URL = 'https://es.wikipedia.org'
```

```python
# Firefox web driver path
# Download the driver for you S.O. here: https://github.com/mozilla/geckodriver/releases
DRIVER_PATH = '/BigData/web-scraping-presidentes/geckodriver.exe'
```

```python
# Creating a new firefox window
#from selenium import webdriver

# Creating a new firefox window
browser = webdriver.Firefox()
#browser = webdriver.Firefox(executable_path = DRIVER_PATH, options=options, firefox_profile=firefox_profile)
```

```python
def make_request(browser, relative_path):
    # Making the request and rendering the browser
    browser.get(SITE_URL + relative_path)
    
    # Simulating vertical scrolling for handling lazy load
    check_height = browser.execute_script('return document.body.scrollHeight;')
    while True:
        browser.execute_script('window.scrollTo(0, document.body.scrollHeight);')
        time.sleep(3)
        height = browser.execute_script('return document.body.scrollHeight;')
        if height == check_height: 
            break 
        check_height = height
    
    # Getting HTML content and passing it to BeautifulSoup for scraping analysis
    return BeautifulSoup(browser.page_source, 'html.parser')
```

```python
# Getting HTML content for news listing page
soup = make_request(browser, '/wiki/Anexo:Presidentes_de_Colombia')
```

```python
# Finding the section where news are contained 
layout = []
layout = soup.find_all('table')
#print(layout)
```

```python
# Getting rows and columns from layout
presidents = []
#row search
del layout[0]
for s in range(len(layout)):
    rows = layout[s].find_all('tr') #find rows
    del rows[0]
    for r in rows:
        colums = r.find_all('td') #find columns
        presidents.append([
            colums[1].get_text().replace('\n',''),
            #colums[2].get_image().replace('\n',''),
            colums[3].get_text().replace('\n',''),
            colums[4].get_text().replace('\n','')
            
        ])
```
